**matplotlib**
- matplotlib is converts the raw data into graphical representation.
- linear plot
- bar graph
- histogram
- scattered 
- pie chart
- box plot
- subgraph

```python
import matplotlib.pyplot as plt
```

```python
plt.plot([12,14,2,90,3],[4,76,23,89,5])
plt.xlabel("x_axis")
plt.ylabel("y_axis")
plt.title("linear graph")
plt.show()
```

```python
print(dir(plt),end="")
```

```python
plt.plot([12,14,2,90,3],[4,76,23,89,5],"g-*",label="safe zone")
plt.plot([3,23,5,45],[98,67,6,34],color="red",label="danger zone")
plt.plot([1,3,2,4],[10,20,30,1],label="n")
plt.legend()
plt.xlabel("x_axis",color="red")
plt.ylabel("y_axis",color="blue")
plt.title("multi linear graph",color="green")
plt.grid()
plt.show()
```

**subplots**
- to create multiple plots in the same figure.

```python
import numpy as np
x = np.linspace(45,90,30)
y = np.log(x)


plt.figure(1)


plt.subplot(2,2,1)
plt.title("log(x)")
plt.grid()
plt.plot(x,y)

plt.subplot(2,2,2)
plt.title("log(x)**2")
plt.grid()
plt.plot(x,y*2)

plt.figure(2)

plt.subplot(2,2,1)
plt.title("log(x)**3")
plt.grid()
plt.plot(x,y**3)

plt.subplot(2,2,2)
plt.title("log(x)**4")
plt.grid()
plt.plot(x,y**4)

plt.show()
```
